```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#img1
img1 = cv2.imread('image1.jpeg')
```

```python
#img1.shape
```

```python
#img2
img2 = cv2.imread('image2.jpeg')
```

```python
#img2.shape
```

```python
dim=(1000,600)
resized_img1 = cv2.resize(img1, dim, interpolation = cv2.INTER_AREA)
resized_img2 = cv2.resize(img2, dim, interpolation = cv2.INTER_AREA)
```

```python
# resized_img1.shape
# resized_img2.shape
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

```python
blend = cv2.addWeighted(resized_img1, 0.6, resized_img2, 0.6, 0.0)
```

```python
cv2.imwrite('blended.png', blend)
```

```python
plt.imshow(blend)
```
